##### 4.1.3. (5 คะแนน) Is there any row in the data set with all NaNs?  Design your Python script to answer this question. Students must provide comments along your code to explain your solution.
```
Expected Output:
Number of rows with all NaNs = ..........................
```

[use the following code area for answering this question]

````python
# Q.4.1.3.
print("Number of rows with all NaNs =",booking.isnull().all(axis=1).sum()) #ตรวจสอบทุกตัว(all())ใน row นั้นว่าข้อมูลในทุกคอลัมน์ของ row นั้นเป็น Nan หรือไม่และใช้ sum() บวกจำนวนของ row ที่ทุกcolumnเป็น NaN
````

##### 4.1.4. (7 คะแนน) Fill missing values in every column (except the 'country' and 'children' column) with ``0``.
Show how you verify that all NaN values in the specified columns have been replaced.

Students must provide comments along your code to explain your solution.  

[use the following code area for answering this question]

````python
# Q.4.1.4.
for cols in booking.columns: #loop to check all columns in data set
  if cols!= 'country' and cols!='children': #condition to check columns except country 'and' children
    booking[cols].fillna(0, inplace=True) #if column!= country and !=children , fill Nan/missing value with zero.
booking[cols].isnull().any().sum()
booking.info() # print info() to verify that all NaN values in the specified columns have been replaced โดยดูที่จำนวน Non-Nulll count


````

##### 4.1.5. (6 คะแนน) Fill all missing values in column 'country' with the most-frequently occurred (``mode``) value of the column.

Show how you verify that all NaN values in the column 'country' have been replaced.

Students must provide comments along your code to explain your solution.

[use the following code area for answering this question]


````python
# Q.4.1.5.
import statistics  #import statistics to use mode()
booking['country']=booking['country'].fillna(value=statistics.mode(booking['country']))  #replace NaN/missing values by mode(the most-frequently occurred) of country
booking.info() # look at Non-Null count of country to check that all Nan values in column 'country have been replaced , so country = 119390 non-null = ถูกแทนค่าเรียบร้อยแล้ว
````

#####4.1.6. (6 คะแนน) Fill all missing values in column 'children' with the rounded ``mean`` value of the column.

````python
# Q.4.1.6.
booking['children']=booking['children'].fillna(value = booking['children'].mean()) #fill boking['children'].mean() replace missing value/Nan
booking.info() #check Non-Null count of column children ถ้า children = 119390 non-null = ถูกแทนค่าเรียบร้อยแล้ว
````

###6. (23 คะแนน) Data Validity (More -- Challenge)

There might be some rows in the data set that there are children or baby guests without adults. These type of rows does not make any sense.

Find booking rows with such condition and remove all of them.

#####6.1.(18 คะแนน) Identify how many rows in the data set that have children or babies guests, but zero adults.

```
Expected Output:
Number of rows with children or babies guests, but zero adults = .............
```

````python
#Q.6.1.
print("Number of rows with children or babies guests, but zero adults =", booking[(booking['adults']==0) & ((booking['children']!=0) | (booking['babies']!=0))].shape[0])
````

#####6.2.(5 คะแนน) Drop all rows in the data set that have children or babies guests, but zero adults.

Verify also that all rows with total 0 guests have been removed.

Students must provide comments along your code to explain your solution.  

````python
#Q.6.2.
booking=booking.drop(booking[(booking['adults']==0) & ((booking['children']!=0) | (booking['babies']!=0))].index, axis=0) #ดรอปโดยการใช้ index
booking.reset_index(drop=True)
booking[(booking['adults']==0) & ((booking['children']!=0) | (booking['babies']!=0))].shape[0] # เช็คว่าดรอปถูกต้องตามเงื่อนไข ถ้าผลลัธ์มีค่าเป็น 0 แสดงข้อมูลที่ไม่เป็นตามเงื่อนไขถูกลบแล้ว
````
