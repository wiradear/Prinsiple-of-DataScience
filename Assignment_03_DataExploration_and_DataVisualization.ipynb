Data Visualization help to easily understand the complex problem. They help in identifying patterns, relationships, and outliers in data. In addition, data visualization is also an effective and meaningful manner for displaying the results.

In this assignment, We will use the bike rental (day) dataset. The data file ``bike_day.csv`` has been attached to this assignment. Descriptions for every column in the data set has made available with this assignment as follows.
- ``instant`` : record index
- ``dteday`` : date
- ``season`` : season (1:winter, 2:spring, 3:summer, 4:fall)
- ``yr`` : year (0: 2011, 1:2012)
- ``mnth`` : month ( 1 to 12)
- ``holiday`` : weather day is holiday or not (extracted from [Web Link])
- ``weekday`` : day of the week (วันภายในอาทิตย์ 0 = Sun, 1 = Mon, .., 6 = Sat)
- ``workingday`` : if day is neither weekend nor holiday is 1, otherwise is 0.
- ``weathersit`` :
  - ``1``: Clear, Few clouds, Partly cloudy, Partly cloudy
  - ``2``: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
  - ``3``: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
  - ``4``: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- ``temp`` : Normalized temperature in Celsius.
  - The values are derived via (``t`` $-$ ``t_min``)$/$(``t_max`` $-$ ``t_min``), ``t_min``$=$ $-$8, ``t_max``$=$ $+$ 39 (only in hourly scale)
- ``atemp`` : Normalized feeling temperature in Celsius.
  - The values are derived via (``t`` $-$ ``t_min``)$/$(``t_max`` $-$ ``t_min``), ``t_min`` $=$ $-$16, ``t_max`` $=$  $+$50 (only in hourly scale)
- ``hum`` : Normalized humidity. The values are divided to 100 (max)
- ``windspeed`` : Normalized wind speed. The values are divided to 67 (max)
- ``casual`` : count of casual users
- ``registered`` : count of registered users
- ``cnt`` : count of total rental bikes including both casual and registered

````python
#Q.1

#import Python modules as required
import matplotlib as mpt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

#read data set into our working area
day = pd.read_csv('/content/drive/MyDrive/CS245/03_bike_day.csv')

#plt.style.use('default')
#plt.style.use('classic')
plt.xkcd()

fig, ax = plt.subplots(figsize=(12,8))

fontdict={'fontsize': 18,
          'weight' : 'bold',
         'horizontalalignment': 'center'}

fontdictx={'fontsize': 18,
          'weight' : 'bold',
          'horizontalalignment': 'center'}

fontdicty={'fontsize': 16,
          'weight' : 'bold',
          'verticalalignment': 'baseline',
          'horizontalalignment': 'center'}

# weekday=0: Sun, weekday=6 : Sat
# mysize = [(day.weekday+1)*10]  # จะได้ 7 sizes: 1=Sun...7=Sat

mysize = (day.holiday)*100 + (day.workingday)*10
mysize = mysize.replace(0,100)

scatter = ax.scatter('temp', 'cnt', data=day, c='season',s=mysize) #c=colour , s=size

# produce a legend with the unique colors from the scatter
legend_elem = list(scatter.legend_elements())
legend_elem[1] = ['winter', 'spring', 'summer', 'fall/autumn']
legend_season = ax.legend(*legend_elem,
                           loc="upper left", title="Season" )
ax.add_artist(legend_season)

# produce a legend with a cross section of sizes from the scatter
handle_dayofweek, label_dayofweek = scatter.legend_elements(prop="sizes", alpha=0.6)
label_dayofweek = ['working days', 'Holidays']
legend_dayofweek = ax.legend(handle_dayofweek, label_dayofweek, loc="center left", title="Day-of-Week")

ax.set_title('Bike Rentals at Different Temperatures\nBy Season and Working Days/Holidays', fontdict=fontdict, color="black")
ax.set_xlabel("Normalized temperature", fontdict=fontdictx)
ax.set_ylabel("Count of Total Rental Bikes", fontdict=fontdicty);

````

<u>หมายเหตุ</u> เอาท์พุตถูกละไว้ ไม่ได้แสดงทั้งหมด



````python
#Q.7.
dec_fire = df.query('month == "December"').groupby(['state'])['number'].sum()
dec_fire_state = (dec_fire != 0).sum()
group_state = df.groupby(['state'])['number'].sum().count()

print("There are", dec_fire_state, "out of", group_state, "states having fires during December. They are as follows:")
print(df['state'].unique())
````
